````python
# prompt: Prepare the bipartite "club-country", and create two graphs, one where club is projected, and another where country is projected. What is the meaning of the edges in each case?
# Save both projected graphs (weighted) as _graphml_. Include the attributes attributes_country and attributes_club for the nodes that will be projected.

# Create the bipartite "club-country" graph
fifanet2 = nx.Graph()
fifanet2.add_nodes_from(fifa.club, bipartite=0)
fifanet2.add_nodes_from(fifa.nationality, bipartite=1)
edge_list = []
for _, row in fifa.iterrows():
  edge_list.append((row['club'], row['nationality']))
fifanet2.add_edges_from(edge_list)

# Project the graph onto clubs
club_nodes = {n for n, d in fifanet2.nodes(data=True) if d["bipartite"] == 0}
country_nodes = set(fifanet2) - club_nodes

club_projected = nx.bipartite.weighted_projected_graph(fifanet2, club_nodes)

# Add attributes for club nodes
for node in club_projected.nodes():
  club_row = fifa[fifa['club'] == node].iloc[0]
  if 'Champion' in club_row:
    club_projected.nodes[node]['Champion'] = club_row['Champion']


# Project the graph onto countries
country_projected = nx.bipartite.weighted_projected_graph(fifanet2, country_nodes)

# Add attributes for country nodes
for node in country_projected.nodes():
  country_row = fifa[fifa['nationality'] == node].iloc[0]
  if 'Developed' in country_row:
    country_projected.nodes[node]['Developed'] = country_row['Developed']

# Save the projected graphs as graphml
nx.write_graphml(club_projected, "club_projected.graphml")
nx.write_graphml(country_projected, "country_projected.graphml")

# Meaning of edges in projected graphs
# In the club-projected graph, an edge between two clubs means that they share players from the same country
# In the country-projected graph, an edge between two countries means that they have clubs with players from both countries

print("Club-projected graph saved as club_projected.graphml")
print("Country-projected graph saved as country_projected.graphml")
````
